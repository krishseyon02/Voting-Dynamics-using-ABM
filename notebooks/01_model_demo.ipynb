```python
from src.voting_model import VotingABM
model = VotingABM(n=200, graph="ws", p_edge=0.05, influence=0.8, noise=0.02, steps=150, seed=7)
metrics = model.run()
metrics
```

```python
import matplotlib.pyplot as plt
avg = metrics["avg_over_time"]
plt.figure()
plt.plot(range(len(avg)), avg)
plt.xlabel("Step"); plt.ylabel("Average opinion (share of 1s)")
plt.title("Opinion trajectory")
plt.show()
```

```python
import networkx as nx
import numpy as np
final = np.array(model.history[-1])
pos = nx.spring_layout(model.G, seed=0)
nx.draw(model.G, pos, node_size=40, node_color=final, cmap="coolwarm", with_labels=False)
```
