```python
# === Setup and Imports ===
import sys, os
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Fix the path so Python can find your src/ folder ---
notebook_dir = os.path.dirname(os.path.abspath("__file__"))  # current notebook folder
project_root = os.path.abspath(os.path.join(notebook_dir, ".."))  # one level up
if project_root not in sys.path:
    sys.path.append(project_root)

# --- Now import your model ---
from src.voting_model import VotingABM
```

```python
# === Run a Single Simulation ===
model = VotingABM(n=200, p_edge=0.05, influence=0.8, noise=0.02,
                  zealot_frac=0.0, steps=100, graph="ws", seed=42)
metrics = model.run()

print("Consensus:", metrics["consensus"])
print("Final variance:", metrics["final_variance"])
print("Minority share:", metrics["minority_share"])
```

```python
# === Average Opinion Over Time ===
avg = np.array(metrics["avg_over_time"])
plt.figure(figsize=(6,4))
plt.plot(avg, color="navy")
plt.title("Average Opinion Over Time")
plt.xlabel("Step")
plt.ylabel("Mean Opinion (0â€“1)")
plt.grid(True)
plt.show()
```

```python
# === Visualize the Final Network State ===
plt.figure(figsize=(6,6))
G = model.G
pos = nx.spring_layout(G, seed=42)
node_colors = ['#ff7f0e' if o == 1 else '#1f77b4' for o in model.opinions]
nx.draw(G, pos, node_color=node_colors, node_size=50, edge_color='lightgray')
plt.title("Final Opinions on Network (orange=1, blue=0)")
plt.show()
```

```python
# === Compare Influence and Noise Levels ===
configs = [
    (0.8, 0.02),  # high influence, low noise
    (0.8, 0.10),  # high influence, high noise
    (0.2, 0.02),  # low influence, low noise
    (0.2, 0.10)   # low influence, high noise
]

plt.figure(figsize=(8,6))
for inf, noi in configs:
    m = VotingABM(n=200, p_edge=0.05, influence=inf, noise=noi,
                  steps=100, graph="ws", seed=42)
    res = m.run()
    plt.plot(np.array(res["avg_over_time"]),
             label=f"infl={inf}, noise={noi}")

plt.title("Effect of Influence and Noise on Consensus Formation")
plt.xlabel("Step")
plt.ylabel("Average Opinion")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# === Demonstrate Effect of Zealots ===
plt.figure(figsize=(6,4))
for zf in [0.0, 0.05, 0.1]:
    m = VotingABM(n=200, p_edge=0.05, influence=0.8, noise=0.02,
                  zealot_frac=zf, steps=100, graph="ws", seed=42)
    res = m.run()
    plt.plot(np.array(res["avg_over_time"]),
             label=f"zealot_frac={zf}")

plt.title("Impact of Zealots on Consensus Dynamics")
plt.xlabel("Step")
plt.ylabel("Average Opinion")
plt.legend()
plt.grid(True)
plt.show()
```
